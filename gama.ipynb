```
import bs4
import requests
import os
#import selenium
import json
import io
from time import sleep
from random import randint, uniform
from selenium import webdriver
#from pyvirtualdisplay import Display
from kora.selenium import wd
from selenium.webdriver.common.keys import Keys
```

```
class Spider():
    def __init__(self):
		    self.url_to_crawl = 'https://dominio.com/'
		    #self.all_items = []

	# Open headless chromedriver
    def start_driver(self):
        self.driver = wd
        userAgent = "Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36"
        options = webdriver.ChromeOptions()
        options.add_argument(f'user-agent={userAgent}')
        #wd = webdriver.Chrome('chromedriver', options=options)
        #self.driver.maximize_window()
        sleep (4)


	# Close chromedriver
    def close_driver(self):
        #print('closing driver...')
		    #self.display.stop()
        self.driver.quit()
        print('closed')

	# Tell the browser to get a page
    def get_page(self,url):
        s = str(url)
        self.driver.get(s)
        sleep(uniform(2,2.5))


    def grab_pages(self):
        with io.open('/content/drive/MyDrive/RAlinks/story.json', 'r', encoding='utf-8-sig') as data:
            dct = json.load(data)

            for i in dct:
                for j in dct[i]:
                    url=j['url']
                    self.get_page(str(url))
                    sleep(uniform(2,5))
                    #self.driver.find_element_by_xpath('.//div[@class="vp-shade vp-shade-invisible"]').click()
                    #sleep(uniform(2,5))
                    #hgt = self.driver.execute_script('return document.querySelector("article").offsetHeight')
                    vimeo = self.driver.execute_script('return document.getElementsByTagName("iframe")[0].attributes[0].value')
                    j['vimeo'] = vimeo
                    #print(topico + ' printado! Salvo em: ' + assunto + '/' + capitulo + '/')
                    sleep(1)

            with io.open('/content/drive/MyDrive/RAlinks/otp.json', 'w') as otp:
         	    json.dump(dct, otp)


    def login(self):
        #self.driver.find_element_by_xpath('.//button[#text="Accept All"]').click()
        sleep(4)
        #self.driver.save_screenshot('chupaCookies.png')
        self.driver.find_element_by_xpath('.//input[@type="email" and @class="string email required form-control section_right__form_control"]').send_keys('viniciusveloso@poli.ufrj.br')
        self.driver.find_element_by_xpath('.//input[@type="password" and @class="password required form-control section_right__form_control"]').send_keys('senha')
        #sleep(uniform(2,3))
        self.driver.save_screenshot('inputLogin.png')
        sleep(uniform(2,3))
        self.driver.find_element_by_xpath('.//input[@type="submit" and @class="btn btn btn-primary section_right__button"]').click()
        sleep(randint(2,4))
        self.driver.save_screenshot('Logado.png')
        print('Logado!')
        sleep(1)
        self.driver.find_element_by_xpath('.//div[@class="card" and @style="display: flex; align-items: center;"]').click()
        sleep(1)
        self.grab_pages()




    def parse(self):
        self.start_driver()
        sleep(2)
        self.get_page(self.url_to_crawl)
        sleep(3)
        #self.driver.find_element_by_xpath('.//*[@class="login-btn google"]').click()
        #self.driver.find_element_by_xpath('.//*[@class="login-btn facebook"]').click()
        #sleep(10)
        self.driver.save_screenshot('home.png')
        #self.driver.manage().timeouts().pageLoadTimeout(5, TimeUnit.SECONDS)
        sleep(randint(1,2))
        #self.login()
        self.grab_pages()
        self.close_driver()

```
